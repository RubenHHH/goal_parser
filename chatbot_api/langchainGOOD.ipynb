````python
import torch
from langchain import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, GenerationConfig, pipeline
````

````python
import locale
locale.getpreferredencoding = lambda: "UTF-8"
````

TheBloke/Llama-2-7B-GPTQ    TheBloke/Llama-2-13b-Chat-GPTQ

````python
MODEL_NAME = "TheBloke/Llama-2-13b-Chat-GPTQ"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, use_fast=True)
````

````python
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME, torch_dtype=torch.float16, trust_remote_code=True, device_map="auto")
````

````python
generation_config = GenerationConfig.from_pretrained(MODEL_NAME)
generation_config.max_new_tokens = 1024
generation_config.temperature = 0.0001
generation_config.top_p = 0.95
generation_config.do_sample = True
generation_config.repetition_penalty = 1.15
````

````python
text_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    generation_config=generation_config,
)

llm = HuggingFacePipeline(pipeline=text_pipeline, model_kwargs={"temperature": 0})
````

````python
from langchain.output_parsers import ResponseSchema
from langchain.output_parsers import StructuredOutputParser
from langchain.prompts.chat import ChatPromptTemplate
from langchain.chains import LLMChain, SimpleSequentialChain, ConversationChain
from langchain.memory import ConversationBufferMemory,ConversationBufferWindowMemory, ConversationSummaryBufferMemory
````

````python
status_schema = ResponseSchema(name="Answer",
                             description="The answer given, could be YES or NO")
functions_schema = ResponseSchema(name="Functions",   description="Functions that have been identified among those in [parking recommendation, ticket availability, weather checking, event booking], \
                                                    respond with those identified if there are any and False otherwise")
explanantion_schema = ResponseSchema(name="Expl",
                                    description="If the answer field is NO, the explanation is the rest of the text ; False otherwise")

response_schemas = [status_schema,
                    functions_schema,
                    explanantion_schema]

output_parser = StructuredOutputParser.from_response_schemas(response_schemas)
format_instructions = output_parser.get_format_instructions()
````

````python
from langchain import PromptTemplate

topics = ["parking recommendation", "ticket availability", "weather checking", "event booking"]

template = """Answer to the request \
that is delimited by triple backticks, \
if at least one intention from ["parking recommendation", "ticket availability", "weather checking", "event booking"] is identified in the request made, \
then answer in the form of a list of two elements, the first one is YES and the second one the list of intentions identified, /
otherwise, if no function of ["parking recommendation", "ticket availability", "weather checking", "event booking"] has been identified, answer with No and explain what you would like as input,/
that is : context = [This app is used during the street science days in L'Aquila. Please make requests relating to the following topics: [parking advice, ticket availability, weather check, event booking]. So if the request made does not identify any of these, you must notify the user of this and explain what do you expect also with examples, without stating the functions/topics.]
text: ```{text}```
"""
k = " the list also includes the identified logical operators that connect the identified functions to each other; "

prompt1 = PromptTemplate(
    input_variables=["text"],
    template=template)

chain1 = LLMChain(llm=llm, prompt=prompt1)
````

````python
text = "i want to know how to be a better person "
result = chain1.invoke(input = text)['text']
print(result)
````
